```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
inputs = r"C:\Users\DAI.STUDENTSDC\Desktop\NLP_CV\dataset\CV\images"
outputs = r"C:\Users\DAI.STUDENTSDC\Desktop\NLP_CV\CV\outputs"
img = cv2.imread(os.path.join(inputs, 'circle.png'), 0 )
lenna = cv2.imread(os.path.join(inputs, 'lenna.png'), 0 )
```

```python
plt.imshow(img, cmap = 'grey')
```

```python
help(cv2.cornerHarris)
```

```python
corners = cv2.cornerHarris(img, blockSize = 2, ksize = 3, k = .004)
corners
```

```python
plt.imshow(corners, cmap = 'grey')
```

```python
lenna_corner = cv2.cornerHarris(lenna, blockSize = 2, ksize = 3, k = .04)
new = lenna_corner < lenna_corner.max() / 2
plt.imshow(new, cmap = 'gray')
```
