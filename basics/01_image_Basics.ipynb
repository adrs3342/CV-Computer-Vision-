#### create a image 8x8 chess board

```python
chess = np.empty([8,8], dtype = 'uint8')
```

```python
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            chess[i,j] = 0
        else:
            chess[i,j] = 255
plt.imshow(chess, cmap = 'gray')
```
