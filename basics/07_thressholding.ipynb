```python
binary_lenna = lenna > 127

new_lenna = binary_lenna.astype(np.uint8) * 255

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(lenna, cmap = 'grey')
plt.subplot(1,2,2)
plt.title('Binary')
plt.imshow(binary_lenna, cmap = 'grey')
```

#### using `cv2.threshold`

```python
new_lenna = cv2.threshold(lenna, cv2.CV_64F, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(lenna, cmap = 'grey')
plt.subplot(1,2,2)
plt.title('Binary')
plt.imshow(binary_lenna, cmap = 'grey')
```

### inverse binary thresholding

```python
binary_lenna = lenna < 127

new_lenna = binary_lenna.astype(np.uint8) * 255

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(lenna, cmap = 'grey')
plt.subplot(1,2,2)
plt.title('Binary')
plt.imshow(binary_lenna, cmap = 'grey')
```

```python
new_lenna = cv2.threshold(lenna, cv2.CV_64F, 127, 255, cv2.THRESH_BINARY_INV)

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(lenna, cmap = 'grey')
plt.subplot(1,2,2)
plt.title('Binary')
plt.imshow(binary_lenna, cmap = 'grey')
```

## Otsu Thresholding

```python
thresh_otsu, lenna_otsu = cv2.threshold(lenna, 100, 255, cv2.THRESH_OTSU)
thresh_bin, lenna_bin = cv2.threshold(lenna, 100, 255, cv2.THRESH_BINARY)
plt.figure(figsize = (16,9))
plt.subplot(1, 3, 1)
plt.title('Original')
plt.imshow(lenna, cmap = 'grey')
plt.subplot(1,3,2)
plt.title(f'Binary - {thresh_bin}')
plt.imshow(lenna_bin, cmap = 'grey')
plt.subplot(1,3,3)
plt.title(f'otsu - {thresh_otsu}')
plt.imshow(lenna_otsu, cmap = 'grey')
```
