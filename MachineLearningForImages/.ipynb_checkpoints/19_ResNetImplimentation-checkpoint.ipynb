```python
import tensorflow as tf

def identity_block(x, filters):
    """
    Implements the identity block for ResNet.

    Arguments:
    x -- Input tensor of shape (m, height, width, channels)
    filters -- Tuple of integers, the number of filters for each convolution layer (f1, f2)

    Returns:
    Output tensor of the identity block
    """
    f1, f2 = filters
    x_shortcut = x

    # First component of the main path
    x = tf.keras.layers.Conv2D(filters=f1, kernel_size=(3, 3), padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    # Second component of the main path
    x = tf.keras.layers.Conv2D(filters=f2, kernel_size=(3, 3), padding='same', activation=None)(x)
    x = tf.keras.layers.BatchNormalization()(x)

    # Add shortcut to main path
    x = tf.keras.layers.Add()([x, x_shortcut])
    x = tf.keras.layers.Activation('relu')(x)

    return x
```

```python
def convolution_block(x, filters, s=2):
    """
    Implements the convolutional block for ResNet.

    Arguments:
    x -- Input tensor of shape (m, height, width, channels)
    filters -- Tuple of integers, the number of filters for each convolution layer (f1, f2)
    stride -- Integer, the stride for the first convolution layer

    Returns:
    Output tensor of the convolutional block
    """
    f1, f2 = filters
    x_shortcut = x

    # First component of the main path
    x = tf.keras.layers.Conv2D(filters=f1, kernel_size=(3, 3), strides=s, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)

    # Second component of the main path
    x = tf.keras.layers.Conv2D(filters=f2, kernel_size=(3, 3), strides=1, padding='same')(x)
    x = tf.keras.layers.BatchNormalization()(x)


    # Shortcut path
    x_shortcut = tf.keras.layers.Conv2D(filters=f2, kernel_size=(1, 1), strides=s, padding='same')(x_shortcut)
    x_shortcut = tf.keras.layers.BatchNormalization()(x_shortcut)

    # Add shortcut to main path
    x = tf.keras.layers.Add()([x, x_shortcut])
    x = tf.keras.layers.Activation('relu')(x)

    return x
```

### design the resnet architecture

```python
def ResNet(input_shape, num_classes):
    inputs = tf.keras.Input(shape = input_shape)

    # stage-1
    x = tf.keras.layers.Conv2D(64, (5, 5), strides = (2, 2))(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    x = tf.keras.layers.MaxPool2D((3,3), strides = (2, 2))(x)

    # stage-2 
    x = convolution_block(x, filters = [64, 64], s = 1)
    x = identity_block(x, filters = [64, 64])

    # stage- 3
    x = convolution_block(x, filters = [128, 128], s = 1)
    x = identity_block(x, filters = [128, 128])

    # stage- 4
    x = convolution_block(x, filters = [256, 256], s = 1)
    x = identity_block(x, filters = [256, 256])

    # stage- 3
    x = convolution_block(x, filters = [512, 512], s = 1)
    x = identity_block(x, filters = [512, 512])

    x = tf.keras.layers.GlobalAveragePooling2D()(x)

    outputs = tf.keras.layers.Dense(num_classes, activation = 'softmax')(x)
    model = tf.keras.Model(inputs, outputs)
    return model
```

```python
model = ResNet((32, 32, 3), 10)
```

```python
plot_model(model)
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 5)
```
