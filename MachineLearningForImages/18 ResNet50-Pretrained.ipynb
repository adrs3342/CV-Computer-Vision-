```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout, Input, BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input     # taking data from preproccessing
from tensorflow.keras.applications.resnet50 import decode_predictions   # for prediction
```

```python
image = load_img('TEMPLE.webp', target_size=(224, 224))  # size b/c VGG having trained on 224x224
image
```

```python
# image to array
image = img_to_array(image)
```

```python
# model trained on RGB and image should be in RGB(3 channel) either in color or black and white
image = image.reshape(1, 224, 224, 3)
```

```python
# preprocess the image
image = preprocess_input(image)
image
```

```python
model = ResNet50()
```

```python
result = model.predict(image)
```

```python
len(result[0])
```

```python
decode_predictions(result)
```
